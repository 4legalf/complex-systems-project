Processing Ketamine data of healthy participants- 


its shaped similarily to the resting imperial processing, however, the lengths are very different I exluded the ones with the smallest number of samples for now



```python
import os
from scipy.io import loadmat
import numpy as np

# Set folder path
folder = r"C:\Users\s2364956\Desktop\Data\ComplexSystems\Ketamine\Healthy\EFKAM KET EEG resting data"

all_eeg_data = []
filenames = []

# Loop through files
for file in os.listdir(folder):
    if not (file.endswith('.mat') and 'clean' in file.lower()):
        continue

    path = os.path.join(folder, file)
    try:
        mat = loadmat(path, struct_as_record=False, squeeze_me=True)
        
        # Extract data
        data = mat['data_clean']
        trials = data.trial

        trials_data = []

        for trial_idx, trial in enumerate(np.atleast_1d(trials)):
            if trial.size == 0:
                continue

            trial_data = trial.T  # (timepoints, channels)
            trials_data.append(trial_data)

        # Stack all trials along time
        eeg_data_recording = np.vstack(trials_data)
        all_eeg_data.append(eeg_data_recording)
        filenames.append(file)

    except Exception as e:
        print(f"Failed to load {file}: {e}")

# Show number of samples before trimming
print("\nSamples per recording before trimming:")
for i, rec in enumerate(all_eeg_data):
    print(f"{filenames[i]}: {rec.shape[0]} samples")

# Find the two smallest ones
sample_counts = [rec.shape[0] for rec in all_eeg_data]
sorted_indices = np.argsort(sample_counts)
excluded_indices = sorted_indices[:7]

excluded_files = [filenames[i] for i in excluded_indices]
excluded_samples = [sample_counts[i] for i in excluded_indices]

print("\nExcluded recordings:")
for f, s in zip(excluded_files, excluded_samples):
    print(f"{f} ({s} samples)")

# Remove excluded recordings
for idx in sorted(excluded_indices, reverse=True):
    all_eeg_data.pop(idx)
    filenames.pop(idx)

# New minimal samples after exclusion
new_min_samples = min(rec.shape[0] for rec in all_eeg_data)
print(f"\nNew minimum number of samples to trim to: {new_min_samples}")

# Cut all remaining recordings
all_eeg_data_trimmed = [rec[:new_min_samples, :] for rec in all_eeg_data]

# Stack and transpose to (recordings, channels, time)
eeg_3d = np.stack(all_eeg_data_trimmed, axis=0)
eeg_3d = np.transpose(eeg_3d, (0, 2, 1))

print(f"\nFinal EEG Data Shape (Recordings × Channels × Time): {eeg_3d.shape}")
print(f"Each recording has {eeg_3d.shape[2]} samples (timepoints) and {eeg_3d.shape[1]} channels.")

```
